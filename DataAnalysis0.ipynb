# From Random Numbers Generators(useful for simulations for data science experiments)

NumPy's random module provides functions to generate random numbers and arrays.Functions like randint, rand,randn are useful functions to generate random numbers for simulations and experiments in data science.

#rand

Generate random numbers from a uniform distribution over [0,1].

```python
import numpy as np
#Example 1
np.random.rand(4)
```

```python
#Example 2
np.random.rand(50)
```

# randint

Generate random integers from low (inclusive) to high(exlusive)

```python
# Example 1
np.random.randint(1, 6, size = 12 ) # random integers between 1 and 5 (6 excluded)
```

```python
# Example 2:Find seed for reproducibility
np.random.seed(42)

np.random.randint(1, 6, size = 12 ) # random integers between 1 and 5 (6 excluded)
```

# randn
