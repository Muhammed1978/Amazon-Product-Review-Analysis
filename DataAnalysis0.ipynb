#Importing packages with alias

It is common for python users to import pacakages with alias.

Alias helps us to shorten the name of the packages while rferencing a certian function with it.

The syntax looks this:


```python
#Example 1
import module_name as another_name
import numpy as np
```

***We can now refer to the function as *np.function_name *rather *numpy.function_name*

```python
#Example 2 :For the constant(pi) which is rounded to 3.142
numpypi = np.pi
print(numpypi)
```
