(c) *Loading Functions from a package*

To load se;scted functions from a module , we use the following syntax:


```python
from math import sqrt

# Now you can use sqrt directly without the 'math.' prefix
result = sqrt(81)
print(f"The square root of 81 is {result}")
```

In this example:

*   `from math import sqrt` specifically imports only the `sqrt` function from the `math` module.
*   This allows you to call `sqrt()` directly, without having to prefix it with `math.` (e.g., `math.sqrt()`).

This can be useful if you only need a few specific functions from a large module and want to keep your code cleaner.

**You can also use those functions without referencing numpy again.
