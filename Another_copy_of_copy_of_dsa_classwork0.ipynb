EXERCISE:GUESS THE NUMBER

In this exercise,initialise a **variable** with a number between **1 and 20**.Giving the user **6 chances** ,tell the user to **guess your chosen number**.Tell the user if his/her **chosen number is greater or less than** your **chosen number** or if he/she **guessed your chosen number right.**

````python
# solution
# initialize the variable with a random number between 1 and 20
chosen_number = 8

# set the number of chances to 6
num_chances = 6

# loop for 6 chances
for i in range(num_chances):

    # get the user's guess
    user_guess = int(input(f"Guess my number between 1 and 20 (Chances left: {num_chances - i}): "))

    # check if the guess is correct
    if user_guess == chosen_number:
        print("Congratulations! You guessed it right.")
        break

    # check if the guess is too high
    elif user_guess > chosen_number:
        print("Your guess is too high.")

    # check if the guess is too low
    else:
        print("Your guess is too low.")

# If the loop finishes without a correct guess
if user_guess != chosen_number:
    print(f"Sorry, you've run out of chances. The correct number was {chosen_number}.")
````

### Difference between List and Tuple Assignment

The fundamental difference between lists and tuples in Python is their mutability:

*   **Lists are mutable**: This means that after a list is created, you can change its elements, add new elements, remove existing elements, or even assign new values to specific indices or slices of the list.
*   **Tuples are immutable**: Once a tuple is created, its elements cannot be changed. You cannot modify an element, add or remove elements, or assign new values to a slice of a tuple. If you try to, Python will raise a `TypeError`.

````python
# Example of List Assignment
my_list = [1, 2, 3, 4, 5]
print(f"Original list: {my_list}")

# Modifying an element in a list
my_list[0] = 10
print(f"List after modifying an element: {my_list}")

# Assigning to a slice of a list
my_list[1:3] = [20, 30]
print(f"List after assigning to a slice: {my_list}")

# Adding elements to a list (not direct assignment, but demonstrates mutability)
my_list.append(60)
print(f"List after appending an element: {my_list}")

print("\n" + "-"*30 + "\n")

# Example of Tuple Assignment (attempted and resulting in error)
my_tuple = (1, 2, 3, 4, 5)
print(f"Original tuple: {my_tuple}")

# Attempting to modify an element in a tuple will raise a TypeError
try:
    my_tuple[0] = 10
except TypeError as e:
    print(f"Error when trying to modify an element in a tuple: {e}")

# Attempting to assign to a slice of a tuple will also raise a TypeError
try:
    my_tuple[1:3] = (20, 30)
except TypeError as e:
    print(f"Error when trying to assign to a slice of a tuple: {e}")

# To 'change' a tuple, you must create a new one
new_tuple = my_tuple + (6,)
print(f"New tuple created by concatenation: {new_tuple}")

# Or convert to a list, modify, and convert back
temp_list = list(my_tuple)
temp_list[0] = 100
modified_tuple = tuple(temp_list)
print(f"Tuple modified via list conversion: {modified_tuple}")
````

**Sets**

. Sets contain unique elements, i.e no appeals are allowed.

. The elements in a set do not have an order

.Sets cannot contain elements which can be internally modified (e.g list and dictionaries)


````python
l = [1,2,3,2,3] # list of 5 values
s = set(l) # set of 3 uniques values
print(s)
e = set() # empty set
print(e)

number = {5,6,8,9,6}
print(number)

{1,2,3}
set()
{8,9,5,6}
````

***Sets are very similar to lists,and tuples,you can use many of the same operators and functions, except they are inherently unordered, so they don't have an index, and can only contain uniques values,so adding a value already in the set will have no effect.

````python
#Example
s = set([1,2,3,2,3,])
print(s)
print("number in set:", len(s))
s.add(4)
print(s)
s.add(3)
print(s)
````

***You can remove specific elements from the set

````python
s = set([1,2,3,2,3])
print(s)
s.remove(3)
print(s)

````

You can do all the expected logical operations on sets,such as taking the union or intersection of 2 sets with the | or and & and operators.

````python
s1 = set ([2,4,6,8,10])
s2 = set ([4,5,6,7])

print("union:",s1 | s2)
print ("intersection:", s1 & s2)
````

# **Dictionaries**

For these situations, a dictionary is a useful data structure.

  . Dictionary do not share data in any particular order

  . The first thing you see dictionary is open and close curly brackets.
