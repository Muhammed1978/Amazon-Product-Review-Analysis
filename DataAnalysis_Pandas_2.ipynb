# Working with missing data

In this section we will discuss values also referred to as **NA**, **NaN** or **NULL** in pandas.

When exploring data you'll most likely encounter missing values, which symbolically represent non-existence values.

We can check whether there are missing values in one or more columns of a DataFrame by using *.isna()*or *.isnull()* while *.isna().sum()* or *.isnull().sum() can be used to get the number of missing values in each of the columns of DataFrame.

We check whether there is missing value by using followings:

* *.isna()*

 or

* *.isnull()*

# Data Imputation

 **Strategies**

There are two options in dealing with missing values:

1. Get rid of rows or columns with nulls

2. Replace nulls with non-null values which is known as data imputation.

3. Machine Learning



# Strategy 1

# 1. Getting rid of rows or columns with nulls

 * Every data scientist or analyst regularly face the problem of whether to remove (drop) the missing values(NA, NULL, or NAN) or imputate them.

 * This is a decision that requires intimate knowlegde of your data and its content.Overall, removing null data is only suggested if you have a small amount of missing data.

 * Removing nulls is pretty simple by using *.dropna()*.


# Example 1

For example let's remove the missing values in the Promotion dataset.Remember in our Promotion_Dataset DataFrame, there are some missing values in qualification column.


```python
# Step 1: Check the head of promotion dataset
Promotion_Dataset.head()
```

Let's use *.shape* to check the number of rows and columns before removing any missing values.

```python
# Step 2: Check the shape of Promotion_Dataset;
Promotion_Dataset.shape
```

 Now let's drop those missing values *.dropna() attribute.This action will delete any row with at least a single NA, Null, or NAN value.

 But it will return a new DataFrame without altering the original one.You could specify *inplace=True* in this method as well.

```python
# Step 3: Using drop attribute for Promotion_Dataset
Promotion_Dataset.dropna()
```

**Note**

* You will notice that prior to using *drop attiribute*, the shape of the promotion dataset was *38312 rows and 19 columns*.

* After using the *drop attibute* the data shape change to *36633 rows and 19 columns*.In which  *1679 rows* were dropped due to missing values.

* Also, other than just dropping rows, you can drop columns with null values by seting axis = "columns" or 1

**Important Note**

Sometimes there can be *UnicodeDecodeError* when reading CSV file in Pandas with Python .You will notice that I  used *encoding = "150-8859-1"* option to stop this error.
