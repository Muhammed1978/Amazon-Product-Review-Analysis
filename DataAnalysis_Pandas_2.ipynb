* Excel workbook can have many worksheets and can  host different data .while importing your data,you need to specify the index of the sheet that you want to import.

* Remember Python index start from 0.

```python
# Example: sheet_names = 0, =>it will print 1st sheet from excel page
east_africa_names = pd.read.excel("datasets/east-Africa-names.xlxs",  sheet_name= 0)
east_africa_names
```

**Note**

* while importing excel data using above code, you alterantively write specify the *sheet_name = sheet1* instead of *sheet_name = 0*

 * it will be observed that *sheet_name = 0*, means that *sheet 1* from the excel page will be dispalyed,meanwhile if the *sheet_name = 1*,the second sheet is displayed.

Therefore, it is necessary to specify the sheet name or number to avoid error.
