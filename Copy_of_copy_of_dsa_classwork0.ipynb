WHILE LOOP

Generally,the while loop is used to keep looping through a block of code while the condition is True. It can also be used to loop through a block of code a certain number of times as seen in the code below:

#while condition and the increment +1


````python
#Example 2
x = 5 #initialize our iterator x
while (x <= 10): #while is for looping
     print(x)
     x+= 1  #increment
````

AS seen in the example above,in a while loop,we need to have initialised a variable before the while statement ,in the case of our example 'a' was installed to 5.Another mportant thing to note is that the variable must be incremented to prevent the loop from executing forever(infinite loop).

THE CONTINUE STATEMENT

The** continue** statement is used to stop the current iteration and move on to the next iteration.When the program execution reaches a continue statement,the program execution immediately jumps back to the start of the loop and re-evaluates the loop's condition.

Suppose we want to print  odd numbers between 1 and 50,we can use a continue statement to achieve it.


````python
#Example
a = 0 #initialization
while ( a <= 50):
     a += 1
     if(a % 2 == 0):
          continue #all the next iteration
     print(a)
````

EXERCISE:GUESS THE NUMBER

In this exercise,initialise a **variable** with a number between **1 and 20**.Giving the user **6 chances** ,tell the user to **guess your chosen number**.Tell the user if his/her **chosen number is greater or less than** your **chosen number** or if he/she **guessed your chosen number right.**

````python
# solution
# initialize the variable with a random number between 1 and 20
chosen_number = 8

# set the number of chances to 6
num_chances = 6

# loop for 6 chances
for i in range(num_chances):

    # get the user's guess
    user_guess = int(input(f"Guess my number between 1 and 20 (Chances left: {num_chances - i}): "))

    # check if the guess is correct
    if user_guess == chosen_number:
        print("Congratulations! You guessed it right.")
        break

    # check if the guess is too high
    elif user_guess > chosen_number:
        print("Your guess is too high.")

    # check if the guess is too low
    else:
        print("Your guess is too low.")

# If the loop finishes without a correct guess
if user_guess != chosen_number:
    print(f"Sorry, you've run out of chances. The correct number was {chosen_number}.")
````
