# Pandas DataFrame

* A DataFrame represents a rectangular table of data and contains an ordered collection of columns,each of which can be a different value type (integer, float, boolean,etc).

* We can think of DataFrame as a bunch of series of objects put togther to share the same index.

# How to create a Pandas DataFrame

There are many ways to create a DataFrame scratch but a great option is to create it from a dictionary.

# Creating DataFrame from a Dictionary

* Among the many things that can serves as input to make a Pandas DataFrame is a  dict (Dictionay).

* To make a data frame from a dictionary, you will need to pass it to the DataFrame function in the data argument.


# Example 1

DataFrame

 Gender        Country         Age

    Male        Uganda          11

    Female     Ethiopia         17

    Female     Rwanda           15

    Female     Somalia          17

    Male       Sudan            13

    Female     Kenya            11

    Male       Uganda            9

    Female     Kenya            18

    Male       Somalia          13

    Female     Rwanda           16   

```python
# Example: Creating DataFrame
registrations_dict = {"Gender": ["Male", "Female", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
                      "Country": ["Uganda", "Ethiopia", "Rwanda", "Somalia", "Sudan", "Kenya", "Uganda", "kenya", "Somalia", "Rwanda"],
                      "Age":[11, 17, 15, 17, 13, 11, 9, 18, 13, 16]}

registrations_dict
```

```python
type(registrations_dict)
```

And then pass the result to the pandas Dataframe constructor.

```python
import pandas as pd
registrations_df = pd.DataFrame(registrations_dict)
registrations_df
```

**The resulting DataFrame will have its index assigned automatically as with Series, and the columns are placed unordered .if you specify a sequence of columns, the DataFrame's will be arranged in that order.**

```python
registrations_df = pd.DataFrame(registrations_dict, columns = ["Country", "Gender", "Age",])
registrations_df
```

# Viewing Your Data

For a large DatFrames, we can select the first five observations or rows by using *.head()* attributes and the last five observation by using *.tail()* attribute.

**Note**

The essence of using *.head() and .tail()* attributes to have easy access to your data,imagine you wan to view 2 million data set at once, this will ctashe your computere.

```python
import pandas as pd
registrations_dict = {"Gender": ["Male", "Female", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
                      "Country": ["Uganda", "Ethiopia", "Rwanda", "Somalia", "Sudan", "Kenya", "Uganda", "kenya", "Somalia", "Rwanda"],
                      "Age":[11, 17, 15, 17, 13, 11, 9, 18, 13, 16]}
registrations_df = pd.DataFrame(registrations_dict)
registrations_df
```
