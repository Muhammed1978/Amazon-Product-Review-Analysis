# Selection with .loc and .iloc attribute

* The loc[] and iloc[] DataFrame attribute enable you to select a subset of the rows and columns from a DataFrame with NumPy -like notation using either axis labels *(df.loc[])* or *integers* *(df.iloc[])*.

* The general format for both *loc[] and iloc[]* are *df.loc[row, col]* where *df* is the name of your DataFrame while the *[row, col]* specifies the row and column index of the dataFrame.

* Remember Python counting starts at 0, so the first row is row zero.

# .loc[]

As a preliminary example, let's select first three rows and columns such as Country and Population by using *df.loc[]*.

# Example 1

*df.loc[row, col]* is for selection by column label.

```python
COVID19data_df
```

```python
COVID19data_df.loc[0:3, ["Country", "Population"]]
```

```python
COVID19data_df.loc[ : , ["Country", "Total Deaths", "Population"] ]
```

In example 1, we selected the first three observations and variables Country and Population.

**Important note**

Please note that *.loc* index position is inclusive i.e 0:5 means 0, 1, 2, 3, 4, 5. This is different from NumPy array that 0 : 5 is 0, 1 , 2, 3, 4.

# Example 2

# Conditional Filtering of a DataFrame

In our regstration_df DataFrame,we can show a DataFrame where only Female are available.

```python
import pandas as pd
registrations_dict = {"Sex":["Male", "Female", "Female", "Female", "Male", "Female","Male", "Female", "Male", "Female"],
                      "Country":["Uganda", "Ethiopia", "Rwanda", "Somalia", "Sudan", "Kenya", "Uganda", "Kenya", "Somalia", "Rwanda"],
                      "Age":[11, 17, 15, 17, 13, 11, 9, 18, 13, 16 ]}

registrations_df = pd.DataFrame(registrations_dict)
registrations_df
```

# Methodology

**Step 1:**
Select Gender Series from the DataFrame

```python
step1 = registrations_df["Sex"]
step1
```

**Step2:**

Use comparison == operator to distinguish whether they are "Female" or not.

```python
Step2 = registrations_df["Sex"] == "Female"
Step2
```

**Step 3:**

Use the result of step 2 to select /filter the registrations_df DataFrame.

We do step 1 to 3 in one line of code as shown below:

```python
registrations_df[registrations_df["Sex"] == "Female"]
```

# Example 2

In the COVID19data_df DataFrame, show a DataFrame where:

1. COVID-19 Cases is less than 3000

* The resulting DataFrame has how many rows?

2. COVID-19 Death Cases is less than 50 and Total Recovered is 200.

* The resulting DataFrame has many rows?

```python
import pandas as pd
COVID19data = { "Country": ["Burundi","Ethiopia","Kenya","Rwanda","Somalia","Tanzania","Uganda","Sudan"],
               "Total Cases": [673, 107109, 79322, 5750, 4445, 509, 18406, 16431],
                "Total Deaths": [1, 1664, 1417, 47, 113, 21, 186, 1202],
                "Total Recovered": [575, 66575, 52974, 5241, 3412, 183, 8764, 9854,],
                "Active Cases": [97, 38871, 24931, 462, 920, 305, 9456, 5375],
                "Population": [1202914, 116082175, 54237961, 13078334, 16066789, 60399786, 46304101, 44252054]}
COVID19data_df = pd.DataFrame(COVID19data)
COVID19data_df.rename(columns = {"Total Cases": "COVID-19 Cases",
                                 "Total Deaths": "COVID-19 Death Cases"}, inplace = True)
COVID19data_df["Discharge rate"] = COVID19data_df["Total Recovered"]/COVID19data_df["COVID-19 Cases"]
COVID19data_df
```

```python
# Solution 1: COVID-19 Cases < 3000
COVID19data_df[COVID19data_df["COVID-19 Cases"] < 3000]
```

# Solution 1

* COVID-19 Cases is less than 3000

  The resulting DtaFrame has how many rows?




```python
# Solution 1:  The resulting DataFrame has how many rows?
COVID19data_solution1 = COVID19data_df[COVID19data_df["COVID-19 Cases"] < 3000]
COVID19data_solution1.shape
```

## Solution 2

-COVID-19 Death Cases is less than $50$ and Total $200$
  
  The resulting DataFrame has how many rows?

As we can from the question that the condition is based on two Series/Columns

1. COVID-19 Death Cases is less than $50$

2. Total Recovered is less than $200$



```python
# Solution2: COVID-19 Death Cases < 50, & Total recovere < 200
COVID19data_df_solution2 = COVID19data_df[(COVID19data_df["COVID-19 Death Cases"] < 50) & (COVID19data_df["Total Recovered"] < 200)]
```
